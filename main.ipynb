```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Replace 'your_dataset.csv' with the actual name of your dataset file
df = pd.read_csv('water_potability.csv')
```

```python
df.head()
df.info()
```

```python
# Example: Filtering data
filtered_df = df[df['ph'] > 1]
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Hardness', y='Solids', data=filtered_df)
plt.title('Bar Plot')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Solids'], bins=20, kde=True)
plt.title('Histogram')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Sulfate', y='Chloramines', data=filtered_df)
plt.title('Scatter Plot')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Hardness', y='Conductivity', data=filtered_df)
plt.title('Box Plot')
plt.show()
```
